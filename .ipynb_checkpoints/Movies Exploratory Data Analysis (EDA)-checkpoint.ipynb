```python
def searchList(given_string,search_list):
    given_list = given_string.split(',')
    match_case = 0
    search_list = re.sub(' +', ' ', search_list.title()).replace(", ",",").split(",")
    found = 0
    for i in search_list:
        if i != "":
            #print(i)
            if i in given_list:
                match_case += 1
                found += 1
            else:
                found = 0
                return 0
                break
        else:
            continue
    if found == len(search_list): return 1
```

```python
def searchList(given_string, search_list):
    given_list = given_string.split(',')
    search_list = re.sub(' +', ' ', search_list.title()).replace(", ",",").split(",")
    
    if all(i in given_list for i in search_list):
        return 1
    else:
        return 0
```
